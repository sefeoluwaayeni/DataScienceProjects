**What is the average score of students who completed the test preparation course?**

```python
# filter the dataset to only include students who completed the test preparation course 
complete_df = new_data.loc[new_data['Test preparation course'] == 'Completed']

# calculate the average score of completed students
avg_score = complete_df['Total_Score'].mean()

# print the average score
print('The average score of students who completed the test preparation course is:', avg_score)
```

**What is the average score of students who didnt take the test preparation course?**

```python
# filter the dataset to only include students who completed the test preparation course students
none_df = new_data.loc[new_data['Test preparation course'] == 'None']

# calculate the average score of completed students
avg_score = none_df['Total_Score'].mean()

# print the average score
print('The average score of students who didnt take the test preparation course is:', avg_score)
```

The average score of the students who completed the test preparation course is higher than those who didnt take the test preparation course. This can indicate that the data points in the high mean value group are generally higher or greater in some way than those in the low mean value group which means that the students who took the test preparation course did better.

**Was the lunch time clashing with the time for the test preparation course?**

```python
#get_dummies in pandas takes information in a table and separates it into different categorie
#check = pd.get_dummies(new_data[['Test preparation course']])
```

```python
df = new_data[(new_data["Lunch"] == "Yes") & (new_data["Test preparation course"] == "Completed")]
```

```python
num = df["Student ID"].count()
print("The number of students who went for lunch and still completed the test preparation course is",num)
```

```python
df1 = new_data[(new_data["Lunch"] == "No") & (new_data["Test preparation course"] == "None")]
num1 = df1["Student ID"].count()
print("The number of students who didnt go for lunch and didnt complete the test preparation course is",num1)
```

```python
df2 = new_data[(new_data["Lunch"] == "Yes") & (new_data["Test preparation course"] == "None")]
num2 = df2["Student ID"].count()
print("The number of students who went for lunch and didnt complete the test preparation course is",num2)
```

```python
df3 = new_data[(new_data["Lunch"] == "No") & (new_data["Test preparation course"] == "Completed")]
num3 = df3["Student ID"].count()
print("The number of students who didnt go for lunch but they completed the test preparation course is",num3)
```

```python
values = [num, num1, num2, num3]
```

```python
fig, ax = plot.subplots()
ax.bar(range(len(values)),values)
ax.set_xlabel('Value index')
ax.set_ylabel('Value')
ax.set_title('Variation in values')
plot.show()
```

It can be deduced that the lunch time was not clashing with the test preparation course time as the number of students who had lunch and still completed their course is greater than all other factors
